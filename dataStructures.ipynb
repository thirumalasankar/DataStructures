```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
    def insertAtEnd(self,node):
        if self.head is None:
            self.head=node
        else:
            lastNode=self.head
            while True:
                if lastNode.next is None:
                    break
                lastNode=lastNode.next
            lastNode.next=node
    def insertAtBegin(self,node):
        if not self.head:
            self.head = node
        tempNode = self.head
        self.head = node
        self.head.next = tempNode
        del tempNode
    def insertAtPosition(self,node,pos):
        c = 1
        if pos == 1:
            self.insertAtBegin(node)
            return 
        cur = self.head
        while(c != pos - 1):          #1-->2-->3-->4-->n
            curNode = curNode.next
            c += 1
        tmp = cur.next
        cur.next = node
        node.next = tmp
    def deleteAtEnd(self):
        if not self.head:
            print("Empty Linked list")
        lstNode = self.head
        while lstNode.next.next:
            lstNode = lstNode.next
        lstNode.next = None
    def deleteAtBegin(self):
        if not self.head:
            print("Empty Linked List")
        self.head = self.head.next
    def deleteAtPosition(self,pos):
        cur = self.head
        c = 1
        if pos == 1:
            self.deleteAtBegin()
            return
        while c != pos-1:
            cur = cur.next
            c += 1 
        cur.next = cur.next.next
    def reverse(self):
        prev,cur = None,self.head
        while cur:
            temp = cur.next    # 1--2--3--null
            cur.next = prev
            prev = cur
            cur = temp
        self.head = prev
        self.printList()
    def printList(self):
        currentNode=self.head
        while True:
            if currentNode is None:
                break;
            print(f"{currentNode.data}",end="-->")
            currentNode=currentNode.next
        print("Null")
n=int(input())
ll=LinkedList()
for _ in range(n):
    node=Node(input())
    ll.insertAtEnd(node)
ll.insertAtBegin(Node("0"))
ll.insertAtPosition(Node(1),2)
ll.printList()
ll.deleteAtEnd()
ll.printList()
ll.deleteAtBegin()
ll.printList()
ll.deleteAtPosition(3)
ll.printList()
ll.reverse()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.prev=None
        self.next=None
class DoublyLinkedList:
    def __init__(self):
        self.head=None
    def insertAtBegin(self,newNode):
        if self.head is None:
            self.head=newNode
        else:                             #2-->3-->Null
            tempNode=self.head              
            self.head=newNode  
            self.head.next=tempNode
            tempNode.prev=self.head
            del tempNode
    def insertAtPosition(self,newNode,pos):
        currentPosition=0
        currentNode=self.head
        while (pos!=currentPosition):
            previousNode = currentNode
            currentNode = currentNode.next
            currentPosition+=1
        previousNode.next=newNode
        newNode.prev=previousNode
        newNode.next=currentNode
        currentNode.prev=newNode
    def insertAtEnd(self,newNode):
        if self.head is None:
            self.head=newNode
        else:
            lastNode=self.head
            while lastNode.next is not None:
                lastNode=lastNode.next
            lastNode.next=newNode
            newNode.prev=lastNode
    def deleteAtBegin(self):
        if self.head is None:
            print("empty")
        else:
            fnode=self.head
            snode=self.head.next
            self.head=snode
            snode.prev=None
            del fnode
    def deleteAtPosition(self,pos):
        currentNode=self.head
        currentPos=0
        while currentPos!=pos:
            previousNode=currentNode
            currentNode=currentNode.next
            currentPos+=1
        previousNode.next=currentNode.next
        currentNode.next.prev=previousNode
    def deleteAtEnd(self):
        if self.head is None:
            print("empty")
        else:
            lastNode=self.head
            while lastNode.next.next is not None:
                lastNode=lastNode.next
            lastNode.next=None
    def forwardTraversal(self):
        if self.head is None:
            print("empty")
        else:
            currentNode=self.head
            while currentNode:
                print(currentNode.data,end="-->")
                currentNode=currentNode.next
            print("Null")
    def backwardTraversal(self):
        if self.head is None:
            print("empty")
        else:
            lastNode=self.head
            while lastNode.next is not None:
                lastNode=lastNode.next
            currentNode=lastNode
            while currentNode:
                print(currentNode.data,end="-->")
                currentNode=currentNode.prev   
            print("Null")
n1=Node(1)
n2=Node(2)
n3=Node(3)
d1=DoublyLinkedList()
d1.insertAtEnd(n1)
d1.insertAtEnd(n2)
d1.insertAtEnd(n3)
d1.insertAtBegin(Node(0))
d1.insertAtPosition(Node("hi"),2)
for data in range(11,15,1):
    d1.insertAtBegin(Node(data))
d1.forwardTraversal()
#d1.backwardTraversal()
d1.deleteAtEnd()
d1.deleteAtBegin()
d1.forwardTraversal()
#d1.backwardTraversal()
d1.deleteAtPosition(1)
d1.deleteAtEnd()
d1.forwardTraversal()
```

```python
class bsTree:
    def __init__(self,key):
        self.key=key
        self.lchild=None
        self.rchild=None
    def insert(self,data):
        if self.key is None:
            self.key = data
            return
        if self.key==data:
            return
        if data < self.key:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild=bsTree(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild=bsTree(data)
    def search(self,data):
        if self.key==data:
            print("present")
            return 
        if data < self.key:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("not found")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("not found")
    def preorder(self):
        print(self.key,end=" ")
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()
    def inorder(self):
        if self.lchild:
            self.lchild.inorder()
        print(self.key,end=" ")
        if self.rchild:
            self.rchild.inorder()
    def postorder(self):
        if self.lchild:
            self.lchild.postorder()
        if self.rchild:
            self.rchild.postorder()
        print(self.key,end=" ")
    def delete(self,data):
        if self.key==None:
            print("empty tree")
        if self.search(data)=="present":
            pass
        else:
            print("not found")
root=bsTree(10)
l=[6,3,1,6,98,3,7]
for i in l:
    root.insert(i)
#print(root.search(sele))
print("this is preorder")
root.preorder()
print()
print("this is inorder")
root.inorder()
print()
print("this is postorder")
root.postorder()
delele=7
root.delete(delele)
```

```python
#all permutations
def permutations(l):
    size = len(l)
    if size == 0:
        yield []
    elif size == 1:
        yield l
    else:
        for i in range(size):
            ele = l[i]
            x = l[ :i] + l[i+1: ]
            for p in permutations(x):
                yield [ele]+p
arr=[1,3,2]
for lis in permutations(arr):
    print(*lis)
```
